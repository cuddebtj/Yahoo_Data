```python
import pandas as pd
pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_rows", None)
```

```python
teams = pd.DataFrame()
for i in range(2013, 2021):
    t = pd.read_csv(f"data/teams/league-teams-{str(i)}.csv")
    teams = pd.concat([teams, t])
teams.replace(["Shinn", "ryan", "Wesley", "Peter Billups"],["James", "Ryan", "Wes", "Pete"], inplace=True)
teams = teams[teams.Manager != "--hidden--"]
teams = teams[teams.Manager != "Brian"]
teams = teams[teams.Manager != "Jake"]
teams.drop(["Points_For", "Points_Against", "Team_Key", "Trades", "Transactions"], axis=1, inplace=True)
encoded_grades = {"A+": 4, "A": 4, "A-": 3.7, "B+": 3.3, "B": 3.0, "B-": 2.7, "C+": 2.3, "C": 2.0, "C-": 1.7, "D+": 1.3, "D": 1.0}
teams["Draft_Grade_GPA"] = teams["Draft_Grade"].replace(encoded_grades)
teams
```

```python
ovall_rank_gpa = teams.groupby(["Rank"]).mean()
ovall_rank_gpa.drop(["Playoff_Seed", "Year", "Draft_Position"], axis=1, inplace=True)
ovall_rank_gpa.sort_values("Draft_Grade_GPA", ascending=False)
```

```python
manager_gpa = teams.groupby(["Manager"]).mean()
manager_gpa.drop(["Year", "Draft_Position"], axis=1, inplace=True)
manager_gpa.sort_values("Draft_Grade_GPA", ascending=False)
```

```python
draft_gpa = teams.groupby(["Draft_Position"]).mean()
draft_gpa.drop(["Year",], axis=1, inplace=True)
draft_gpa.sort_values("Draft_Grade_GPA", ascending=False)
```
